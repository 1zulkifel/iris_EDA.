# Plotting for explorarity data analysis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
irs_d= pd.read_csv('./irs.csv')
irs_d
irs_d.shape
irs_d.describe
irs_d.columns
```

```python
irs_d['Species'].value_counts()
```

# 2-d Scatter plot

```python
irs_d.plot(kind="scatter",x="SepalLengthCm",y="SepalWidthCm"); 
plt.show()
```

```python
sb.set_style("darkgrid");
sb.FacetGrid(irs_d,hue="Species",size=6).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend();
plt.show()

    
```

# 3-d Scatter plot

```python
plt.close();
sb.set_style("whitegrid");
sb.pairplot(irs_d,hue="Species",size=4,diag_kind="kde");
plt.show()
```

```python
plt.close();
sb.set_style("darkgrid");
sb.pairplot(irs_d,hue="Species",size=3);
plt.show()
```
